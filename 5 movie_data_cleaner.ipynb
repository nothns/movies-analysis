```python
import pandas as pd
import numpy as np
import json
import regex as re
```

```python
df = pd.read_json("data_final.json", encoding="UTF8")
df
```

```python
# Filtered out non-movies
df = df[df['Type']=="movie"]
df
```

```python
# Removed columns
df.drop(['Rated', 'Director', 'Writer', 'Actors', 'Poster', 'imdbID', 'Type', 'DVD', 'Production', 'Website', 'Response', 'totalSeasons'], axis=1, inplace=True)
df
```

```python
# Removed duplicates
df.drop_duplicates(["Title", "Year", "Released", "Runtime"], inplace=True)
df
```

```python
# Converted nulls
df.replace("N/A", np.NaN, inplace=True)
df.replace("nan", np.NaN, inplace=True)
df
```

```python
# Removed movies with limited information
mask = df['imdbVotes'].isnull() | df['imdbRating'].isnull()
df = df[~mask]
df
```

```python
# Fixed basic types and converted runtime
df = df.convert_dtypes()
df["Year"] = df["Year"].astype("Int64")
df["Released"] = pd.to_datetime(df["Released"], format="%d %b %Y")
df["Runtime"] = df["Runtime"].apply(lambda x: x if pd.isna(x) else int(x.split()[0]) if len(x.split()) == 2 else int(x.split()[0])*60+int(x.split()[2]) if len(x.split()) == 4 else np.NaN)
df["Runtime"] = df["Runtime"].astype("Int64")
```

```python
# fix awards

df.columns
df[["Won", "Nominated"]] = np.NaN
df = df[['Title', 'Year', 'Released', 'Runtime', 'Genre', 'Plot', 'Language', 'Country',
       'Awards', 'Won', 'Nominated', 'Ratings', 'Metascore', 'imdbRating', 'imdbVotes',
       'BoxOffice']]

def search_wins(x):
    if pd.isna(x):
        return x
    win_pattern = r"([0-9]+)[^0-9]*win[s]*"
    num = re.search(win_pattern, x, re.IGNORECASE)
    if num:
        return int(num.group(1))
    else:
        return 0

def search_nominations(x):
    if pd.isna(x):
        return x
    nomination_pattern = r"([0-9]+)[^0-9]*nomination[s]*"
    num = re.search(nomination_pattern, x, re.IGNORECASE)
    if num:
        return int(num.group(1))
    else:
        return 0

df["Won"] = df["Awards"].apply(lambda x: search_wins(x))
df["Nominated"] = df["Awards"].apply(lambda x: search_nominations(x))
df.drop(['Awards'], axis=1, inplace=True)
df["Won"] = df["Won"].astype("Int64")
df["Nominated"] = df["Nominated"].astype("Int64")
df[["Won", "Nominated"]].replace(np.NaN, 0, inplace=True)
df
```

```python
# fix ratings
df["TomatoScore"] = np.NaN
df = df = df[['Title', 'Year', 'Released', 'Runtime', 'Genre', 'Plot', 'Language', 'Country',
              'Won', 'Nominated', 'Ratings', 'Metascore', 'TomatoScore', 'imdbRating', 'imdbVotes',
       'BoxOffice']]

def tomatoes(x):
    for rating in x:
        if rating['Source'] == 'Rotten Tomatoes':
            return rating["Value"][:-1]
    return np.NaN

df["TomatoScore"] = df["Ratings"].apply(lambda x: tomatoes(x))
df.drop(['Ratings'], axis=1, inplace=True)
df["Metascore"] = df["Metascore"].astype("Int64")
df["TomatoScore"] = df["TomatoScore"].astype("Int64")
df
```

```python
# fixed imdbRating
df["imdbRating"] = pd.to_numeric(df["imdbRating"])
df.dtypes
```

```python
df["imdbVotes"] = df["imdbVotes"].apply(lambda x: str(x).replace(",", ""))
df["imdbVotes"] = df["imdbVotes"].astype("Int64")
df
```

```python
df["BoxOffice"] = df["BoxOffice"].apply(lambda x: str(x).replace("$", ""))
df["BoxOffice"] = df["BoxOffice"].apply(lambda x: str(x).replace(",", ""))
df.replace("<NA>", np.NAN, inplace=True)
df.replace("nan", np.NAN, inplace=True)
df["BoxOffice"] = df["BoxOffice"].astype("Int64")
df
```
