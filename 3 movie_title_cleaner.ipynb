```python
import numpy as np
import pandas as pd
import sqlite3
import regex as re
```

```python
con = sqlite3.connect("database")
cur = con.cursor()
cur.execute("SELECT * FROM movie_titles")
res = cur.fetchall()
res
```

```python
df = pd.DataFrame(res, columns=["id", "year", "country", "title", "list_id"])
df = df.set_index(["id"])
df["title_clean"] = np.NAN
df["title_bracket"] = np.NAN
df["title_bracket_2"] = np.NAN
df
```

```python
list(df.columns.values)
df = df [['year', 'country', 'title', 'title_clean', 'title_bracket', 'title_bracket_2', 'list_id']]
df
```

```python
pattern = '\((.*)\)'
df['title_bracket'] = df['title'].str.findall(pattern)
# find inside bracket
df['title_bracket'] = df['title_bracket'].apply(lambda x: x[0] if x else np.NaN)
# remove first word inside bracket -- some brackets contain (aka [alternative name]) so need to remove the aka
df['title_bracket_2'] = df['title_bracket'].apply(lambda x: re.match(".*? (.*)", str(x)).group(1) if re.match(".*? (.*)", str(x)) else np.NaN)
# get a clean title that does not have brackets
df['title_clean'] = df['title'].apply(lambda x: x if not re.findall("\(", str(x)) else re.match("(.*)\(", str(x)).group(1))
df
```

```python
df[df["title_clean"].str.contains("\[", na=False)]
# some titles have [] brackets that need to be removed
df['title_clean'] = df['title_clean'].apply(lambda x: re.match("(.*)\[", str(x)).group(1) if re.findall("\[", str(x)) else x)
df
```

```python
df.to_sql('movie_titles_cleaned', con=con, if_exists='append')
```
